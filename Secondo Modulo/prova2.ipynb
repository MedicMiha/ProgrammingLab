```python
import statistics

salari = [98000, 101000, 102000, 99000, 97000]

mean = statistics.mean(salari)
median = statistics.median(salari)
mode = statistics.mode(salari)

print(f'Mean: {mean}')
print(f'Median: {median}')
print(f'Mode: {mode}')
```

```python
import csv

file_path = "./../esercizi/data.csv"
file = open(file_path)
content = file.read()
file.close()

data_lista = []

for riga in csv.reader(content.strip().split('\n')):
    if riga[0] != 'date':
        data_lista.append(riga)
        
print(data_lista)
```
